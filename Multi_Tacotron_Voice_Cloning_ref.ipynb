```
from pathlib import Path
import subprocess
import os
import tqdm
import shutil

speakers_dirs = Path('/content/Voices1.1')
out_dir = Path('/content/output_generated')
Path.mkdir(out_dir, exist_ok=True)
for speaker in speakers_dirs.iterdir():
    if speaker.is_dir():
        Path.mkdir(out_dir / speaker.name, exist_ok=True)
        print(f'Processing speaker: {speaker.name}')
        for idx, text in enumerate(tqdm.tqdm(open('/content/default_texts.txt', 'r').readlines())):
            audio_path = next(speaker.glob('*'))
            #audio_path = Path('/content/Multi-Tacotron-Voice-Cloning/ex.wav')
            bash_command = f'python  demo_cli.py -p {audio_path.resolve()} -t "{text}" --no_sound'
            process = os.system(bash_command)
            #process = subprocess.call(bash_command.split(), stdout=subprocess.PIPE)
            #output, error = process.communicate()
            #print(output)
            #print(error)
            output_path = Path('/content/Multi-Tacotron-Voice-Cloning/demo_output_00.wav')
            shutil.move(str(output_path.resolve()), str((out_dir / speaker.name / (str(idx) + '.wav')).resolve()))
```
